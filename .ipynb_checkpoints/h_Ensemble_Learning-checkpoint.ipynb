##### Í∑∏Î¶º Ï∂úÏ≤ò: grokking-machine-learning(Î£®Ïù¥Ïä§ ÏÑ∏ÎùºÎÖ∏), Rosy Park, Î©îÏù¥ÌîåÏä§ÌÜ†Î¶¨
- Ïñ¥Îñ§ Îç∞Ïù¥ÌÑ∞Ïùò Í∞íÏùÑ ÏòàÏ∏°ÌïúÎã§Í≥† Ìï† Îïå, ÌïòÎÇòÏùò Î™®Îç∏Îßå Í∞ÄÏßÄÍ≥† Í≤∞Í≥ºÎ•º ÎèÑÏ∂úÌï† ÏàòÎèÑ ÏûàÏßÄÎßå,  
Ïó¨Îü¨ Í∞úÏùò Î™®Îç∏ÏùÑ Ï°∞ÌôîÎ°≠Í≤å ÌïôÏäµ(Ensemble Learning)ÏãúÏºú Í∑∏ Î™®Îç∏Îì§Ïùò ÏòàÏ∏° Í≤∞Í≥ºÎì§ÏùÑ Ïù¥Ïö©ÌïúÎã§Î©¥ Îçî Ï†ïÌôïÌïú ÏòàÏ∏°Í∞íÏùÑ Íµ¨Ìï† Ïàò ÏûàÎã§.

#### Î≥¥ÌåÖ(Voting)
- "ÌïòÎÇòÏùò Îç∞Ïù¥ÌÑ∞ ÏÑ∏Ìä∏"Ïóê ÎåÄÌï¥ ÏÑúÎ°ú Îã§Î•∏ ÏïåÍ≥†Î¶¨Ï¶òÏùÑ Í∞ÄÏßÑ Î∂ÑÎ•òÍ∏∞Î•º Í≤∞Ìï©ÌïòÎäî Î∞©ÏãùÏù¥Îã§.
- ÏÑúÎ°ú Îã§Î•∏ Î∂ÑÎ•òÍ∏∞Îì§Ïóê "ÎèôÏùºÌïú Îç∞Ïù¥ÌÑ∞ ÏÑ∏Ìä∏"Î•º Î≥ëÎ†¨Î°ú ÌïôÏäµÌïòÏó¨ ÏòàÏ∏°Í∞íÏùÑ ÎèÑÏ∂úÌïòÍ≥†, Ïù¥Î•º Ìï©ÏÇ∞ÌïòÏó¨ ÏµúÏ¢Ö ÏòàÏ∏°Í∞íÏùÑ ÏÇ∞Ï∂úÌï¥ÎÇ¥Îäî Î∞©ÏãùÏùÑ ÎßêÌïúÎã§.

#### Î∞∞ÍπÖ(Bagging, Bootstrap Aggregation)
- ÌïòÎÇòÏùò Îç∞Ïù¥ÌÑ∞ ÏÑ∏Ìä∏ÏóêÏÑú "Ïó¨Îü¨ Î≤à Ï§ëÎ≥µÏùÑ ÌóàÏö©ÌïòÎ©¥ÏÑú ÌïôÏäµ Îç∞Ïù¥ÌÑ∞ ÏÑ∏Ìä∏Î•º ÎûúÎç§ÌïòÍ≤å ÎΩëÏùÄ Îí§(Bootstrap)"  
ÌïòÎÇòÏùò ÏòàÏ∏°Í∏∞ Ïó¨Îü¨ Í∞úÎ•º Î≥ëÎ†¨Î°ú ÌïôÏäµÏãúÏºú Í≤∞Í≥ºÎ¨ºÏùÑ ÏßëÍ≥Ñ(Aggregration)ÌïòÎäî Î∞©Î≤ïÏù¥Îã§.
- Voting Î∞©ÏãùÍ≥º Îã¨Î¶¨ Í∞ôÏùÄ ÏïåÍ≥†Î¶¨Ï¶òÏùò Î∂ÑÎ•òÍ∏∞Î•º ÏÇ¨Ïö©ÌïòÍ≥† ÌõàÎ†® ÏÑ∏Ìä∏Î•º Î¨¥ÏûëÏúÑÎ°ú Íµ¨ÏÑ±ÌïòÏó¨ Í∞ÅÍ∏∞ Îã§Î•¥Í≤å(ÎèÖÎ¶ΩÏ†ÅÏúºÎ°ú, Î≥ëÎ†¨Î°ú) ÌïôÏäµÏãúÌÇ®Îã§.  
- ÌïôÏäµ Îç∞Ïù¥ÌÑ∞Í∞Ä Ï∂©Î∂ÑÌïòÏßÄ ÏïäÎçîÎùºÎèÑ Ï∂©Î∂ÑÌïú ÌïôÏäµÌö®Í≥ºÎ•º Ï£ºÏñ¥ Í≥ºÏ†ÅÌï©Îì±Ïùò Î¨∏Ï†úÎ•º Ìï¥Í≤∞ÌïòÎäîÎç∞ ÎèÑÏõÄÏùÑ Ï§ÄÎã§.
- Î∞∞ÍπÖÎ∞©ÏãùÏùÑ ÏÇ¨Ïö©Ìïú ÎåÄÌëúÏ†ÅÏù∏ ÏïåÍ≥†Î¶¨Ï¶òÏù¥ Î∞îÎ°ú ÎûúÎç§ Ìè¨Î†àÏä§Ìä∏ ÏïåÍ≥†Î¶¨Ï¶òÏù¥Îã§.

üìåÎ∂ÄÌä∏Ïä§Ìä∏Îû©(bootstrap)ÏùÄ ÌÜµÍ≥ÑÌïôÏóêÏÑú ÏÇ¨Ïö©ÌïòÎäî Ïö©Ïñ¥Î°ú, random samplingÏùÑ Ï†ÅÏö©ÌïòÎäî Î∞©Î≤ïÏùÑ ÏùºÏª´Îäî ÎßêÏù¥Îã§.

#### Î∂ÄÏä§ÌåÖ(Boosting)

```python
import pandas as pd
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()
cancer_df = pd.DataFrame(cancer.data, columns=cancer.feature_names)
cancer_df['target'] = cancer.target
cancer_df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(7, 5))
correlation_matrix = cancer_df.corr()
heatmap = sns.heatmap(correlation_matrix, cmap='Oranges')
heatmap.set_title("Correlation")
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.model_selection import GridSearchCV, train_test_split

# ÌïòÏù¥Ìçº ÌååÎùºÎØ∏ÌÑ∞ ÌäúÎãù
dt_params = {'max_depth': [5, 6, 7], 'min_samples_split': [7, 8, 9]}
svm_params = {'C': [0.001, 0.01, 0.1, 1, 10, 100],
             'gamma': [0.001, 0.01, 0.1, 1, 10, 100],
             'kernel': ['linear', 'rbf']}
knn_params = {'n_neighbors': [3, 5, 7, 9, 11]}

grid_dt_classifier = GridSearchCV(DecisionTreeClassifier(), param_grid=dt_params, cv=10, refit=True, return_train_score=True, n_jobs=4, error_score='raise')
# ÏÜåÌîÑÌä∏ Î≥¥ÌåÖÏóêÏÑúÎäî Í∞Å Í≤∞Ï†ï ÌÅ¥ÎûòÏä§Î≥Ñ ÌôïÎ•†Ïù¥ ÌïÑÏöîÌïòÍ∏∞ ÎïåÎ¨∏Ïóê, SVCÏóê probabilityÎ•º TrueÎ°ú ÌïòÏó¨
# predict_proba()Î•º ÏÇ¨Ïö©Ìï† Ïàò ÏûàÎèÑÎ°ù Ìï¥Ï§ÄÎã§(ÌóàÏùÄÏÉÅ ÎèÑÏõÄ).
grid_svc_classifier = GridSearchCV(SVC(probability=True), param_grid=svm_params, cv=5, refit=True, return_train_score=True, n_jobs=4, error_score='raise')
# KNNÏóêÏÑú FlagÏò§Î•ò Î∞úÏÉù
# Series ÌÉÄÏûÖÏùò ÌõàÎ†® Îç∞Ïù¥ÌÑ∞ÏóêÎäî flags ÏÜçÏÑ±Ïù¥ ÏóÜÍ∏∞ ÎïåÎ¨∏Ïóê, numpyÎ°ú Î≥ÄÍ≤ΩÌïú Îí§ ÌõàÎ†®ÏãúÏºúÏïº ÌïúÎã§.
grid_knn_classifier = GridSearchCV(KNeighborsClassifier(), param_grid=knn_params, cv=10, refit=True, return_train_score=True, n_jobs=4, error_score='raise')

# Í∞úÎ≥Ñ Î™®Îç∏ÏùÑ "ÌïòÎìú" Î≥¥ÌåÖ Í∏∞Î∞òÏùò ÏïôÏÉÅÎ∏î Î™®Îç∏Î°ú Íµ¨ÌòÑÌïú Î∂ÑÎ•òÍ∏∞
# Ïò§Ï∞® ÌñâÎ†¨
# [[40  2]
#  [ 3 69]]
# Ï†ïÌôïÎèÑ: 0.9561, Ï†ïÎ∞ÄÎèÑ: 0.9718, Ïû¨ÌòÑÏú®: 0.9583, F1:0.9650, AUC:0.9554

# voting_classifier = VotingClassifier(estimators=[('DTC', grid_dt_classifier)
#                                                  , ('SVC', grid_svc_classifier)
#                                                  , ('KNN', grid_knn_classifier)]
#                                      , voting='hard')


# Í∞úÎ≥Ñ Î™®Îç∏ÏùÑ "ÏÜåÌîÑÌä∏" Î≥¥ÌåÖ Í∏∞Î∞òÏùò ÏïôÏÉÅÎ∏î Î™®Îç∏Î°ú Íµ¨ÌòÑÌïú Î∂ÑÎ•òÍ∏∞
# Ïò§Ï∞® ÌñâÎ†¨
# [[37  5]
#  [ 1 71]]
# Ï†ïÌôïÎèÑ: 0.9474, Ï†ïÎ∞ÄÎèÑ: 0.9342, Ïû¨ÌòÑÏú®: 0.9861, F1:0.9595, AUC:0.9335

voting_classifier = VotingClassifier(estimators=[('DTC', grid_dt_classifier)
                                                 , ('SVC', grid_svc_classifier)
                                                 , ('KNN', grid_knn_classifier)]
                                     , voting='soft')

# Îç∞Ïù¥ÌÑ∞ ÏÑ∏Ìä∏ Î∂ÑÎ¶¨
features, targets = cancer.data, cancer.target
X_train, X_test, y_train, y_test = train_test_split(features, targets, stratify=targets, test_size=0.2)

# VotingClassifier ÌïôÏäµ/ÏòàÏ∏°/ÌèâÍ∞Ä
voting_classifier.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, precision_score , recall_score , confusion_matrix, ConfusionMatrixDisplay, f1_score, roc_auc_score
# ÌÉÄÍ≤ü Îç∞Ïù¥ÌÑ∞ÏôÄ ÏòàÏ∏° Í∞ùÏ≤¥Î•º Ï†ÑÎã¨Î∞õÎäîÎã§.
def get_evaluation(y_test, prediction, classifier=None, X_test=None):
#     Ïò§Ï∞® ÌñâÎ†¨
    confusion = confusion_matrix(y_test, prediction)
#     Ï†ïÌôïÎèÑ
    accuracy = accuracy_score(y_test , prediction)
#     Ï†ïÎ∞ÄÎèÑ
    precision = precision_score(y_test , prediction)
#     Ïû¨ÌòÑÏú®
    recall = recall_score(y_test , prediction)
#     F1 score
    f1 = f1_score(y_test, prediction)
#     ROC-AUC
    roc_auc = roc_auc_score(y_test, prediction)

    print('Ïò§Ï∞® ÌñâÎ†¨')
    print(confusion)
    print('Ï†ïÌôïÎèÑ: {0:.4f}, Ï†ïÎ∞ÄÎèÑ: {1:.4f}, Ïû¨ÌòÑÏú®: {2:.4f}, F1:{3:.4f}, AUC:{4:.4f}'.format(accuracy , precision ,recall, f1, roc_auc))
    print("#" * 75)
    
    if classifier is not None and  X_test is not None:
        fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8,4))
        titles_options = [("Confusion matrix", None), ("Normalized confusion matrix", "true")]

        for (title, normalize), ax in zip(titles_options, axes.flatten()):
            disp = ConfusionMatrixDisplay.from_estimator(classifier, X_test, y_test, ax=ax, cmap=plt.cm.Blues, normalize=normalize)
            disp.ax_.set_title(title)
        plt.show()
```

```python
prediction = voting_classifier.predict(X_test)
get_evaluation(y_test, prediction, voting_classifier, X_test)
```
